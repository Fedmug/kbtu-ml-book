# Autoencoders

Autoencoders are feedforward neural networks that reconstruct the input signal at the output. They consist of a hidden layer that serves as a code describing the model. Autoencoders are designed in such a way that they cannot precisely copy the input to the output. Typically, they are constrained in the code dimensionality (smaller than the signal dimension) or penalized for activations in the code. The input signal is reconstructed with errors due to encoding losses, but to minimize these errors, the network is forced to learn to select the most important features.

### Autoencoders find broad practical applications in various fields:

**Dimensionality Reduction**

- Autoencoders are used for reducing the dimensionality of data while preserving important characteristics. This is valuable in areas where it is necessary to decrease data volume while retaining information.


**Content Generation**

- In computer vision and graphics, autoencoders are employed for generating new images, as well as for reconstructing and modifying existing ones.

**Data Quality Enhancement**

- In signal processing and audio technologies, autoencoders are applied to improve the quality of sound and image data, filtering out noise and artifacts.

**Anomaly Detection**

- Autoencoders play a crucial role in detecting anomalies in data, such as financial fraud, defects in manufacturing processes, or medical anomalies.

**Pretraining for Classification Tasks**

- Using autoencoders for pretraining can enhance the efficiency of classification models in scenarios with limited amounts of labeled data.

**Data Compression:**
- The ability of autoencoders to extract key features enables efficient data compression while preserving important information.

**Image Enhancement:**
- Autoencoders are employed for noise reduction, resolution enhancement, and overall improvement of image quality.

### **Examples:**

- *File Compression:* Autoencoders are employed in file compression formats, such as JPEG, to store images in a more compact form.
  
- *Video Enhancement:* In cases of frame loss in videos, autoencoders assist in filling the missing moments, making the viewing experience smoother.

Autoencoders consist of two parts: an encoder, denoted as $g$, and a decoder, denoted as $f$. The encoder transforms the input signal into its representation: $h = g(x)$, and the decoder reconstructs the signal from its code: $x = f(h)$.

The autoencoder, by adjusting $f$ and $g$, aims to learn the identity function $x = (x,f(g(x))$, minimizing some error functional.

```{math}
    :label: autoencoders
    L(x',f(g(x))).
```

In this case, the families of encoder function $g$ and decoder function $f$ are somehow constrained to force the autoencoder to select the most important properties of the signal.

# Keras

**Keras** is an extremely practical high-level deep learning library that can be used with *TensorFlow* or *Theano* surfaces. Fundamentally, divisions vanish, joining them together, and models are produced. Models and layers maintain their internal properties after they are constructed, therefore it is highly convenient to be able to train a layer in one model and use it in another.

*Keras* models may be easily integrated into *TensorFlow/Theano* code (like operations on tensors); they offer an easy-to-learn, yet very flexible, learning process.

The handwritten numbers from the **MNIST** set will be our source of data.

Now let's download it:


````python
from keras.datasets import mnist
import numpy as np

(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.astype('float32') / 255.
x_test  = x_test .astype('float32') / 255.
x_train = np.reshape(x_train, (len(x_train), 28, 28, 1))
x_test  = np.reshape(x_test,  (len(x_test),  28, 28, 1))
````

# Contractive Autoencoder (CAE)

To begin with, let's create the simplest compressing autoencoder with a low-dimensional code consisting of two fully connected layers: an encoder and a decoder.

Since the color intensity is normalized to unity, we will use a sigmoid activation for the output layer.

We will write separate models for the encoder, decoder, and the entire autoencoder. To do this, we will create instances of layers and apply them one after another, finally combining them into models.

````python
from keras.layers import Input, Dense, Flatten, Reshape
from keras.models import Model

# Dimension of encoded representation
encoding_dim = 49
````

In order to transfer an image to a neural network, first we need to convert
the pixel matrix into a one-dimensional array

````python
input_img = Input(shape=(28, 28, 1))
flat_img = Flatten()(input_img)
````

Next we define a hidden layer, in which the input image is compressed into
a smaller representation through a dense layer, where encoded_dim is a
variable defining the number of neurons in the layer (i.e. to what size
the image will be compressed) and with an activation function
and ReLu (the most commonly used function activations in deep learning)

````python
encoded = Dense(encoding_dim, activation='relu')(flat_img)
````

After this we create the Decoder. The Decoder is a mirror image of the
Encoder, so we do everything the same, simply substituting the reverse values.

````python
input_encoded = Input(shape=(encoding_dim,))
flat_decoded = Dense(28*28, activation='sigmoid')(input_encoded)
decoded = Reshape((28, 28, 1))(flat_decoded)
````

Let's create and compile the model

````python
encoder = Model(input_img, encoded, name="encoder")
decoder = Model(input_encoded, decoded, name="decoder")
autoencoder = Model(input_img, decoder(encoder(input_img)), name="autoencoder")
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
````

Now let's train our autoencoder

````python
%%capture
autoencoder.fit(x_train, x_train,
                epochs=30,
                batch_size=256,
                shuffle=True,
                validation_data=(x_test, x_test),
                verbose=0)
````

````python
# Number drawing function
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

def plot_digits(*args):
    args = [x.squeeze() for x in args]
    n = min([x.shape[0] for x in args])

    plt.figure(figsize=(2*n, 2*len(args)))
    for j in range(n):
        for i in range(len(args)):
            ax = plt.subplot(len(args), n, i*n + j + 1)
            plt.imshow(args[i][j])
            plt.gray()
            ax.get_xaxis().set_visible(False)
            ax.get_yaxis().set_visible(False)

    plt.show()
````

````python
n = 10

imgs = x_test[:n]
encoded_imgs = encoder.predict(imgs, batch_size=n)
encoded_imgs[0]
````

````python
decoded_imgs = decoder.predict(encoded_imgs, batch_size=n)

plot_digits(imgs, decoded_imgs)
````

````python
decoded_images = autoencoder.predict(imgs)
plot_digits(imgs, decoded_images)
````

## Deep Autoencoder

There is nothing preventing us from creating the same autoencoder but with a larger number of layers. In this case, it will be able to capture more complex nonlinear patterns.

````python
encoding_dim = 49

# encoder
input_img = Input(shape=(28, 28, 1))
flat_img = Flatten()(input_img)
x = Dense(encoding_dim*3, activation='relu')(flat_img)
x = Dense(encoding_dim*2, activation='relu')(x)
encoded = Dense(encoding_dim, activation='linear')(x)

# decoder
input_encoded = Input(shape=(encoding_dim,))
x = Dense(encoding_dim*2, activation='relu')(input_encoded)
x = Dense(encoding_dim*3, activation='relu')(x)
flat_decoded = Dense(28*28, activation='sigmoid')(x)
decoded = Reshape((28, 28, 1))(flat_decoded)

# models
d_encoder = Model(input_img, encoded, name="encoder")
d_decoder = Model(input_encoded, decoded, name="decoder")
d_autoencoder = Model(input_img, decoder(encoder(input_img)), name="autoencoder")

d_autoencoder.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
````

Let's take a look at the summary of our model.

````python
d_autoencoder.summary()
````

The number of parameters has increased more than threefold; let's see if the new model performs better:

````python
%%capture
d_autoencoder.fit(x_train, x_train,
                  epochs=50,
                  batch_size=256,
                  shuffle=True,
                  validation_data=(x_test, x_test),
                  verbose=0)
````

````python
n = 10

encoded_imgs = d_encoder.predict(x_test, batch_size=n)
predicted = d_autoencoder.predict(x_test, batch_size=n)

plot_digits(imgs, predicted)
````
