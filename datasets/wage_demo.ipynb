```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'svg'

wage_df = pd.read_csv("wage.csv")
```

```python
wage_df.head()
```

```python
sb.scatterplot(wage_df, x='age', y='wage', c='b', s=10);
```

```python
def plot_regression(order):
    sb.regplot(wage_df, x="age", y="wage", order=order,
           scatter_kws={"color": "blue", "s": 10}, line_kws={"color": "red"}, 
           ).set_title('Wage regression');
```

```python
plot_regression(1)
```

```python
labels = wage_df["education"].unique()
labels.sort()
plt.figure(figsize=(10, 6))
sb.boxplot(wage_df, x="education", y="wage", order=labels);
```
