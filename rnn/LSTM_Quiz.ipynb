```python
from jupyterquiz import display_quiz
import json
from base64 import b64encode
from matplotlib import pyplot as plt
from matplotlib import image as mpimg

def get_spanned_encoded_q(q, q_name):
    byte_code = b64encode(bytes(json.dumps(q), 'utf8'))
    return f'<span style="display:none" id="{q_name}">{byte_code.decode()}</span>'
```

```python
q1 = [{
    "question": "Suppose your training examples are sentences. Which of the following refers to the $j^{th}$ word in the $i^{th}$ training example?",
    "type": "many_choice",
    "answers": [
        {
            "answer": "$x^{(i) < j > }$",
            "correct": True,
            "feedback": "Correct! We index into the $i^{th}$ row first to get the $i^{th}$ training example (represented by parentheses), then the $j^{th}$ column to get the $j^{th}$ word (represented by brackets)."
        },
        {
            "answer": "$x^{ < i > (j)}$",
            "correct": False,
            "feedback": "Incorrect"
        },
        {
            "answer": "$x^{(j) < i > }$",
            "correct": False,
            "feedback": "No, the solution always exists"
        },
        {
            "answer": "$x^{ < j > (i)}$",
            "correct": False,
            "feedback": "Incorrect"
        },
    ]
}]
```

```python
display_quiz(q1)
```

```python
get_spanned_encoded_q(q1, "q1")
```

```python
display_quiz("#q1")
```
