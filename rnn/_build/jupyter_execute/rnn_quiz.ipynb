```python
from jupyterquiz import display_quiz
import json
from base64 import b64encode

def get_spanned_encoded_q(q, q_name):
    byte_code = b64encode(bytes(json.dumps(q), 'utf8'))
    return f'<span style="display:none" id="{q_name}">{byte_code.decode()}</span>'
```

```python
first_q = [{
    "question": "We run RNN, that processes a time series of length $T = 10$, and each time step has an input vector \
    of size $5$. What is the total number of input parameters for the connections from the input layer to the hidden layer?",
    "type": "numeric",
    "answers": [
        {
            "type": "value",
            "value": 50,
            "correct": True,
            "feedback": "Correct!"
        },
        {
            "type": "default",
            "feedback": "No, thats wrong."
        }
    ]
}]
```

```python
display_quiz(first_q)
```

```python
get_spanned_encoded_q(first_q, "first_q")
```
